# Machine Learning Workflow
- 머신러닝은 6단계의 workflow로 이루어집니다.
  1. Acquisition
  2. Inspection and exploration
  3. Preprocessing and Cleaning
  4. Modeling and Training
  5. Evaluation
  6. Deployment

- 정리하자면 아래와 같습니다.
  1. 데이터를 수집하고
  2. 수집한 데이터를 **점검**하고 **탐색**해서 데이터를 파악합니다. (EDA)
  3. 데이터를 파악한 후, 데이터를 **전처리** 합니다.
  4. 적절한 알고리즘을 이용하여 **학습할 모델을 만들고**, 전처리된 데이터를 이용하여 **모델을 학습**합니다.
    - 이때 데이터는 학습, 검증, 테스트 (6:2:2) 비율로 분할하여 약 60% 데이터만 학습에 사용합니다.
  5. model training이 끝났다면 test data로 성능을 평가합니다.
  6. 평가한 성능이 만족스럽다면, 완성된 모델을 배포합니다.
    - 이때, 모델을 업데이트 해야한다면 다시 1단계로 돌아갑니다.

````python
import numpy as np
````

````python
# ndarray 생성

## vector 생성
vec = np.array([1,2,3,4,5])
## matrix 생성
mat = np.array([[10,20,30],[40,50,60]])
##확인
print(vec)
print("==========")
print(mat)
````

````python
# ndarray 타입, 축의 개수(차원 수), 크기 확인
## 타입 확인
print(type(vec))
print(type(mat))
## 축의 개수(차원) 확인
print(vec.ndim)
print(mat.ndim)
## 크기 확인
print(vec.shape)
print(mat.shape)
````

````python
# ndarray 초기화
## 0으로 초기화된 행렬 생성
zero = np.zeros((2,3)) # zeros안에 생성하고 싶은 행렬의 크기(shape)를 작성
## 1로 초기화된 행렬 생성
one = np.ones((2,3)) # shape
## 원하는 값으로 초기화된 행렬 생성
full = np.full((2,3),100) # shape , value
## 랜덤으로 채워진 행렬 생성
random = np.random.random((2,3)) # shape
## 주대각선 값이 1인 행렬 생성
eye = np.eye(3) # 정방행렬이므로 원하는 shape의 값 1개 작성

## 확인
print("<zero>")
print(zero)
print("=========")
print("<one>")
print(one)
print("=========")
print("<full>")
print(full)
print("=========")
print("<random>")
print(random)
print("=========")
print("<eye>")
print(eye)
print("=========")
````

````python
# np.arange()
## 원하는 범위의 배열을 생성하는 방식

# python의 range처럼 사용 가능 (start=0, end, step=1)
## 1. 0~9까지
zero_to_nine = np.arange(10)
## 2. 10~19까지
ten_to_nineteen = np.arange(10,20)
## 3. 0~19까지 짝수만
even = np.arange(0,20,2)
## 확인
print("<0~9>")
print(zero_to_nine)
print("=======")
print("<10~19>")
print(ten_to_nineteen)
print("=======")
print("<even>")
print(even)
````

````python
# np.reshape()
## vector, matrix를 다른 형태로 변환
## 1. zero_to_nine을 2x5으로 변환
zero_to_nine_ver2= np.reshape(zero_to_nine,(2,5)) ## 방식 1 : np.reshape(array,shape)
## 2. zero_to_nine을 5x2로 변환
zero_to_nine_ver3= zero_to_nine.reshape((5,2)) ## 방식 2 : array.reshape(shape)

## 확인
print("<ver2> (2x5)")
print(zero_to_nine_ver2)
print("===============")
print("<ver3> (5x2)")
print(zero_to_nine_ver3)
print("===============")
````

````python
# Numpy Slicing
## python list처럼 slicing이 가능합니다.

## 1. 1행 모든 열 출력
print("1행 모든 열")
print(zero_to_nine_ver2[0,:])
## 2. 2행 마지막 2개 열 출력
print("2행 마지막 2개 열")
print(zero_to_nine_ver2[1,-2:])
## 3. 1-2행 모든 열 출력
print("1-2행 모든 열")
print(zero_to_nine_ver3[:3,:])
## 4. 전체 행의 마지막 열 출력
print("전체 행의 마지막 열")
print(zero_to_nine_ver2[:,-1])
````

````python
# 정수 인덱싱
## ver2의 1,3 위치 값
print("ver2의 <1,3> 위치 값은")
print(zero_to_nine_ver2[0,2])

## 특정 위치의 원소를 가져와 새로운 배열
print("ver2의 1행 3열 값과 2행 1열 값을 가져와 만든 배열은")
new_ver = zero_to_nine_ver2[[0,1],[2,0]]
print(new_ver)
````

````python
# Numpy 연산

## +
add1 = zero_to_nine + ten_to_nineteen
add2 = np.add(zero_to_nine,ten_to_nineteen)
## -
sub1 = zero_to_nine - ten_to_nineteen
sub2 = np.subtract(zero_to_nine,ten_to_nineteen)
## *
mul1 = zero_to_nine * ten_to_nineteen
mul2 = np.multiply(zero_to_nine,ten_to_nineteen)
## /
div1 = zero_to_nine / ten_to_nineteen
div2 = np.divide(zero_to_nine,ten_to_nineteen)

print("0~9와 10~19를")
print("더한 결과")
print(add1)
print(add2)
print("++++++++++++++")
print("뺀 결과")
print(sub1)
print(sub2)
print("++++++++++++++")
print("곱한 결과")
print(mul1)
print(mul2)
print("++++++++++++++")
print("나눈 결과")
print(div1)
print(div2)
print("++++++++++++++")
````

# Data Split
- Split하는 경우
  1. 모델 학습을 진행하기 위해서는 Train,Validation,Test로 분리를 해야 가지고 있는 데이터로 정확하게 훈련부터 모델 평가까지 할 수 있습니다.
  2. 데이터 자체가 X,Y(target)를 모두 가진 경우 Y(target)을 분리해주어야 합니다.
- Split 방식
  1. zip() 이용
    - 2번 경우에 사용할 수 있습니다.
  2. DataFrame() 이용
    - 2번 경우에 사용할 수 있습니다.
  3. numpy slicing 이용
    - 2번 경우에 사용할 수 있습니다.
  4. scikit-learn 이용
    - 1번 경우에 사용할 수 있습니다.
  5. 수동 분리
    - 모든 경우에 사용할 수 있습니다.

**숨겨진 셀을 누르면 코드가 나옵니다.**


````python
#zip() 이용
## zip을 사용하면 동일한 개수를 가지는 sequence의 각 순서별로 묶어서 반환합니다.
X,Y = zip(['a',1],['b',2],['c',3])
print("1차원에 zip 이용")
print("X = ",X)
print("Y = ",Y)
## 이것을 이용하여 아래와 같은 데이터 프레임도 만들 수 있습니다.
df = pd.DataFrame([X,Y])
print("++++++++++++")
print("Zip 결과 DataFrame")
print(df)
print("++++++++++++")
## 아래와 같은 2차원 리스트를 이용할 때는 unpacking 연산자(*)을 이용하여 zip으로 묶을 수도 있습니다.
arr = [['a',1],['b',2],['c',3]]
X,Y = zip(*arr)
print("2차원 리스트에 zip  이용")
print("X = ",X)
print("Y = ",Y)
````

````python
# DataFrame()이용
values = [['a',1],
          ['b',2],
          ['c',3]]
columns = [['알파벳','숫자(Y)']]
df = pd.DataFrame(values,columns = columns)
print("DataFrame",df)
print("++++++++++++")
# 이렇게 하면 알파벳과 숫자를 분리하여 추출할 수 있습니다.
print("X",df["알파벳"])
print("++++++++++++")
print("Y",df["숫자(Y)"])
````

````python
# Numpy를 이용하여 분리하기
array = np.arange(0,16).reshape((4,4))
X = array[:,:3]
Y = array[:,-1]
print(X,Y)
````

````python
# scikit-learn 을 이용
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
print(X_train)
print(X_test)
````

````python
# 수동으로 분리하기
## 1. train_test 분리
X,Y = np.arange(16).reshape(4,4), list(range(4))
## 길이 계산 전체의 x%
rate = 25 #25%를 테스트로 사용 예정
test_len = int(len(X)*rate/100)
train_len = len(X)-test_len
print("test data size :",test_len)
print("train data size :",train_len)
X_train = X[:train_len]
X_test = X[:test_len]
Y_train = Y[:train_len]
Y_test = Y[:test_len]
print(X_train)
print(X_test)
````
